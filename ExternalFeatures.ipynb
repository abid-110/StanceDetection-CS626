```
FinalTrainSet = pd.read_csv('/content/gdrive/MyDrive/StanceDetectionData/FinalTrainSet.csv')
FinalTestSet = pd.read_csv('/content/gdrive/MyDrive/StanceDetectionData/FinalTestSet.csv')
labels_done_train = np.load('/content/gdrive/MyDrive/StanceDetectionData/labels_done_train.npy')
labels_done_test = np.load('/content/gdrive/MyDrive/StanceDetectionData/labels_done_test.npy')
```

```
import csv
train_body_sentences = []
train_headline_sentences = []
test_body_sentences = []
test_headline_sentences = []
with open('/content/gdrive/MyDrive/StanceDetectionData/FinalTrainSet.csv', newline='') as csvfile:
  rows = csv.reader(csvfile, delimiter=',')
  i = -1
  k = 0
  for row in rows:
    print(i)
    if(i==labels_done_train[k]):
      train_body_sentences.append(row[2])
      train_headline_sentences.append(row[1])
      k+=1
      if(k==len(labels_done_train)):
        break
    i+=1
```

```
with open('/content/gdrive/MyDrive/StanceDetectionData/FinalTestSet.csv', newline='') as csvfile:
  rows = csv.reader(csvfile, delimiter=',')
  i = -1
  k = 0
  for row in rows:
    print(i)
    if(i==labels_done_test[k]):
      test_body_sentences.append(row[2])
      test_headline_sentences.append(row[1])
      k+=1
      if(k==len(labels_done_test)):
        break
    i+=1
```

```
#Create external feature vectors for train set

train_ext = []
i = 0
for sent1,sent2 in zip(train_headline_sentences,train_body_sentences):
  print(i)
  i+=1
  vec = []
  #character ngrams
  for n in range(2,17):
    n_grams_1 = ngrams(sent1.lower(), n)
    n_grams_2 = ngrams(sent2.lower(),n)
    vec.append(len(list(set(n_grams_1).intersection(n_grams_2))))

  #word ngrams
  for n in range(2,7):
    n_grams_1 = ngrams(sent1.lower().split(), n)
    n_grams_2 = ngrams(sent2.lower().split(),n)
    vec.append(len(list(set(n_grams_1).intersection(n_grams_2))))

  #Sentence polarity
  flag=False
  text1 = Text(sent1)
  text2 = Text(sent2)
  pol1 = 0
  pol2 = 0
  for word in text1.words:
    try:
      pol1+=word.polarity
    except:
      flag=True
      vec.append(0)
      break
  if (flag==True):
    train_ext.append(vec)
    continue
  flag=False
  for word in text2.words:
    try:
      pol2+=word.polarity
    except:
      flag=True
      vec.append(0)
      break
  if (flag==True):
    train_ext.append(vec)
    continue
  
  pol1 = pol1/(len(sent1.split())*1.0)
  pol2 = pol2/(len(sent2.split())*1.0)
  print(pol1-pol2)
  vec.append(pol1-pol2)

  train_ext.append(vec)


```

```
train_ext = np.array(train_ext)
train_ext.shape
```

```
#Create external feature vectors for train set

test_ext = []
i = 0
for sent1,sent2 in zip(test_headline_sentences,test_body_sentences):
  print(i)
  i+=1
  vec = []
  #character ngrams
  for n in range(2,17):
    n_grams_1 = ngrams(sent1.lower(), n)
    n_grams_2 = ngrams(sent2.lower(),n)
    vec.append(len(list(set(n_grams_1).intersection(n_grams_2))))

  #word ngrams
  for n in range(2,7):
    n_grams_1 = ngrams(sent1.lower().split(), n)
    n_grams_2 = ngrams(sent2.lower().split(),n)
    vec.append(len(list(set(n_grams_1).intersection(n_grams_2))))

  #Sentence polarity
  flag=False
  text1 = Text(sent1)
  text2 = Text(sent2)
  pol1 = 0
  pol2 = 0
  for word in text1.words:
    try:
      pol1+=word.polarity
    except:
      flag=True
      vec.append(0)
      break
  if (flag==True):
    test_ext.append(vec)
    continue
  flag=False
  for word in text2.words:
    try:
      pol2+=word.polarity
    except:
      flag=True
      vec.append(0)
      break
  if (flag==True):
    test_ext.append(vec)
    continue
  
  pol1 = pol1/(len(sent1.split())*1.0)
  pol2 = pol2/(len(sent2.split())*1.0)
  print(pol1-pol2)
  vec.append(pol1-pol2)

  test_ext.append(vec)
```

```
test_ext = np.array(test_ext)
test_ext.shape
```
