```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
train_set = pd.read_csv(".\\Data\\FinalTrainSet.csv")
headline = train_set['Headline']
body = train_set['Body']
stance = train_set['Stance']
headline_wordcount = train_set['Headline Word Count']
body_wordcount = train_set['Body Word Count']
```

```python
print("Mean Length of sentences for headline is "+str(np.mean(headline_wordcount)))
print("Mean Length of sentences for Body is "+str(np.mean(body_wordcount)))
```

```python
# Plot for Headline
plt.title("Headline Word Count")
plt.xlabel("Sentence Word count")
plt.ylabel("Frequency")
plt.hist(headline_wordcount,bins = 10)
plt.show()
```

```python
#plot for body
plt.title("Body Word Count")
plt.xlabel("Sentence Word count")
plt.ylabel("Frequency")
plt.hist(body_wordcount,bins = [0,500,1000,1500,2000,2500])
plt.show()
```

```python
print(Counter(stance).keys()) 
print(Counter(stance).values())
```
